```python
df = df.rename(columns={'消費日': 'Date',
                        '交易說明': 'Details',
                        '新臺幣金額': 'NTD_Amount',
                        '消費國家': 'Country',
                        '幣別': 'Currency',
                        '外幣金額': 'Foreign_Amount'})
df.head()
```

```python
type(df)
```

```python
df.columns
```

```python
# List of columns you want to drop
columns_to_drop = ['入帳起息日', '卡號後四碼', '折算日']

# Drop the columns you don't need
df = df.drop(columns=columns_to_drop, axis=1)
```

```python
df.isna().sum()
```

```python
df.loc[df['Country'].isna()]
```

```python
df['Details'] = df['Details'].replace(' ', '')
df.head()
```

```python
df = df.drop(df.index[:3])
df = df.reset_index(drop=True)
df
```

```python
from pandas.tseries.offsets import DateOffset

# condition = (df.index <= 192) & (df['Date'].dt.year == 2024)

# df.loc[condition, 'Date'] = df.loc[condition, 'Date'] - DateOffset(years=1)

grouped_2023df = df.iloc[:190].groupby('Date')
df.loc[:189, 'Date'] = grouped_2023df['Date'].transform(lambda x: x - DateOffset(years=1))
df
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df.head()
```

```python
df['Category'] = 'unassigned'
df
```

```python
df['Category'] = np.where(df['Details'].str.contains(
    '救國團|運動|中心|樂活',na=False, case=False), 
    'Sports', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '中油|加油站', na=False, case=False), 
    'Gas', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '全聯|楓康|愛買|市場|家樂福|bakery', na=False, case=False), 
    'Groceries', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '中友百貨|餐廳|餐飲|食館|咖哩',na=False, case=False), 
    'Restaurants', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '肯德基|麥當勞|BURGERKING|KFC|McDonald|必勝客|達美樂|Pizza', na=False, case=False),
    'Junk Food', df['Category'])
df['Category'] = np.where(df['Details'].str.contains(
    '優食|Uber',na=False, case=False), 
    'Delivery', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '蝦皮|網購|樂購|momo',na=False, case=False), 
    'Online Shopping', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    'GU|優衣庫|天地',na=False, case=False), 
    'Clothes', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '居家|家居',na=False, case=False), 
    'Furnitures', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    'Nintendo|國外交易手續費|Steam|NINTENDO|STEAM|GAMES|COGNOSPHERE|GOOGLE|APPLE',
    na=False, case=False), 
    'Digital Games', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '手機|LINE|3C',na=False, case=False), 
    'Miscellaneous', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '旅|宿|高鐵|台鐵|航',na=False, case=False), 
    'Travels', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '租車|優步',na=False, case=False), 
    'Car Rental', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '保險|產險',na=False, case=False), 
    'Insurance', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '教育|學校|藝術',na=False, case=False), 
    'Education', df['Category'] )
df['Category'] = np.where(df['Details'].str.contains(
    '點數折抵',na=False, case=False), 
    'Points Returned', df['Category'] )
```

```python
df['Details'].replace({'連加覓靜拾光':'覓靜拾光餐廳', 
                      '1bakert':'1bakery',
                      '愛進化科技股份有限公司':'RhinoShield手機殼',
                      'JAI宅水湳店':'宅餐廳',
                      '輕井澤-新崇德':'輕井澤餐廳',
                      '湯棧-中清店':'湯棧餐廳',
                      '北村豆腐家-台中文心分':'北村豆腐家餐廳',
                      '刁民酸菜魚-公益店':'刁民酸菜魚餐廳'},
                      inplace=True)
```
